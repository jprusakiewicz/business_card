```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_old = pd.read_csv("/Users/kuba/ocr_tests/output/pl/timing_1584615898.246783.csv")
data_new = pd.read_csv("/Users/kuba/ocr_tests/output/pl/timing_1584962847.706655.csv")
```

```python
comparision = data_old.merge(data_new, on="file", suffixes=["_old", "_new"])
```

```python
comparision
```

```python
plt.figure(figsize=(15,8))
plt.bar(comparision.index, comparision.time_old, alpha=0.5, label='old')
plt.bar(comparision.index, comparision.time_new, alpha=0.5, label='new')
plt.xlabel("file")
plt.ylabel("time in seconds")
plt.xticks(np.arange(len(comparision)), comparision.file, rotation=90)
plt.legend()
plt.title("Comparision of capture time between old and new version for each pdf")
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.hist([comparision.time_old, comparision.time_new], bins = 100, density=False, histtype='stepfilled')
plt.title("Histograms of both version time")
plt.show()
```

```python
time_old_df = pd.DataFrame(comparision['time_old'])
time_old_df.columns = ["time"]
time_old_df["version"] = 'old'
time_new_df = pd.DataFrame(comparision['time_new'])
time_new_df.columns = ["time"]
time_new_df["version"] = 'new'
time_both_df = pd.concat([time_new_df, time_old_df], ignore_index=True)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(x="version", y="time", data = time_both_df)
sns.stripplot(x="version", y="time", data = time_both_df, size=8, jitter=True, edgecolor="gray", linewidth=2)
plt.title("Distributions - median, quartiles, minimum and maximum")
plt.show()
```
